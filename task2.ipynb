SINCE THE TEXT CONSISTS MANY UNWANTED THINGS SUCH AS NUMBERS,PUNCTUATION ,TAGS,ALPHANUMERICS AND KEEPING THEM LEADS TO AN UNCLEAN DATA WHICH BECOMES DIFFICULT FOR THE MACHINE TO UNDERSTAND AND GET TRAINED THEREBY REMOVING THESE UNWANTED PROPERTES WE CONVERT THESE  RAW TEXT INTO A PROPER CLEAN DATA WHICH CAN BE USED FOR TRAINING AND TESTING PURPOSE AND IT CAN BE DONE BYBUSING GENSIM MODULE


```python
import gensim
from gensim.parsing.preprocessing import strip_numeric 
from gensim.parsing.preprocessing import strip_non_alphanum
from gensim.parsing.preprocessing import strip_punctuation
from gensim.parsing.preprocessing import strip_punctuation2
from gensim.parsing.preprocessing import strip_multiple_whitespaces
from gensim.parsing.preprocessing import strip_tags
from gensim.parsing.preprocessing import stem_text
from gensim.parsing.preprocessing import preprocess_string
```

```python
filters=[strip_numeric,strip_multiple_whitespaces,strip_non_alphanum,strip_punctuation,strip_punctuation,strip_punctuation2,strip_tags,stem_text]
```

CLEANING THE DATA 

```python
def clean(string):
  words=preprocess_string(string,filters)
  return " ".join(word.lower() for word in words)

```

```python
cl=[]
for i in ns["title"]:
  cl.append(clean(i))
ns["clean"]=cl 
ck=[]
for i in ks["reviews.text"]:
  ck.append(clean(i))
ks["clean"]=ck
```

SINCE WE ARE DOING ANALYSIS FOR TWO SENTIMENTS WE CN CONVERT THE RATING INT0 0 i.e NEGATIVE AND 1 i.e positive 

```python
t=[]
for x in ks["reviews.rating"]:
    if x==5.0:
        t.append(0)
    else:
        t.append(1)
ks["reviews.rating"]=t
ytest=ks["reviews.rating"]
```

```python
gs=ns.drop("title",axis=1)
```

```python
y=[]
for x in gs["rating"]:
    if x>4:
        y.append(1)
    else:
      y.append(0)
gs["rating"]=y
gs
```

## TRAIN AND TESTING OF THE MODEL

THE ALGORITHM USED HERE IS RANDOMFOREST THE REASON BEHIND IT IS THAT IT GAVE A BETTER ACCURACY RATE THAN OTHER ALGORITHMS. THE
