```
import numpy as np
import pandas as pd
df=pd.read_csv("sies.csv",encoding="ISO-8859-1")
df
```

```
k=df.drop(["downvote "," tags"],axis=1)
```

Removing all columns with  missing values and unncecessary data This will allow us to work faster (and our data set is large enough) that it will still be meaningful without them.

```
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(min_df=3,max_features=None,ngram_range=(1,3),stop_words='english',strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}')
```

**Tfidfvectorizer computes the word counts, IDF values, and Tf-idf scores all using the same dataset**.

```
con=tf.fit_transform(k["content "])
```

```
con
```

```
con.shape
```

**Thus data in columns is converted into a sparse matrix**

```
from sklearn.metrics.pairwise import sigmoid_kernel
```

The Hyperbolic Tangent Kernel is also known as the Sigmoid Kernel and as the Multilayer Perceptron (MLP) kernel. The Sigmoid Kernel comes from the Neural Networks field, where the bipolar sigmoid function is often used as an activation function for artificial neurons.
It is used to bring values of entries between 0 and 1.

```
sig=sigmoid_kernel(con,con)
```

```
sig[0]
```

```
tite=pd.Series(k.index,index=k["title "]).drop_duplicates()
```

```
tite
```

```
def grec(title,sig=sig):
    idx=tite[title]
    sc=list(enumerate(sig[idx]))
    sc=sorted(sc,key=lambda x:x[1],reverse=True)
    sc=sc[1:5]
    movie=[i[0] for i in sc]
    return k["title "].iloc[movie]
```

```
grec("SRM #11 - Report")
```
